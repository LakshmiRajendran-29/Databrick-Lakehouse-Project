## To built Bronze Product Information Table

##### Spark Read Raw Product Information from Landing CSV file and Create a new DataFrame


```python
from pyspark.sql import functions as F

base_path = '/Volumes/dlh/landing_db/raw_sales/source_crm/prd_info.csv'

read_data_from_prodinfo = (
  spark.read
    .option("header","true")
    .option("inferSchema", "true")
    .csv(base_path)
    .withColumn("file_path", F.col("_metadata.file_path"))
    .withColumn("ingest_ts" , F.current_timestamp())
)
read_data_from_prodinfo.display()
```

##### Write Product Information from DF to bronzre_prod_info as Delta table 

```python
(read_data_from_prodinfo
        .write.mode('overwrite')
        .format('delta')
        .saveAsTable('dlh.bronze_db.bronze_prod_info'))
```
