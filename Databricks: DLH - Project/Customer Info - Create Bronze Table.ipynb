###01: Customer information

##### Spark Read Raw Customer Information from Landing CSV file and Create a new DataFrame


```python
from pyspark.sql import functions as F

base_path = "/Volumes/dlh/landing_db/raw_sales/source_crm/cust_info.csv"

read_data_from_custinfo = (
    spark.read
    .option('header','true')
    .option('inferSchema','true')
    .csv(base_path)
    .withColumn("file_path", F.col("_metadata.file_path"))
    .withColumn("ingest_ts", F.current_timestamp())
    )
read_data_from_custinfo.display()
```

##### Write Customer Information from DF to bronzre_cust as Delta table 

```python
(read_data_from_custinfo
        .drop(F.col("file_path"))
        .write.mode("overwrite")
        .format("delta")
        .saveAsTable("dlh.bronze_db.bronze_cust_info"))
```
