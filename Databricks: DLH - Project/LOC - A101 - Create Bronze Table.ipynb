## Build Bronze Location A101 Table

##### Spark Read Raw location Information from Landing CSV file and Create a new DataFrame

```python
from pyspark.sql import functions as F

base_path = '/Volumes/dlh/landing_db/raw_sales/source_erp/LOC_A101.csv'

read_data_from_loca101 = (
        spark.read
        .option("header","true")
        .option("inferSchema","true")
        .csv(base_path)
        .withColumn("file_path", F.col("_metadata.file_path"))
        .withColumn("ingest_ts", F.current_timestamp())
)
```

##### Write Customer Information from DF to bronzre_loc_a101 as Delta table 


```python
(read_data_from_loca101
        .write.mode("overwrite")
        .format('delta')
        .saveAsTable('dlh.bronze_db.bronze_loc_a101'))
```
