## Build bronze Sales Details Table

##### Spark Read Raw Sales details from Landing CSV file and Create a new DataFrame

```python
from pyspark.sql import functions as F

base_path = '/Volumes/dlh/landing_db/raw_sales/source_crm/sales_details.csv'

read_data_from_salesdetails = (
  spark.read
  .option("header","true")
  .option("inferSchema","true")
  .csv(base_path)
  .withColumn("file_path", F.col("_metadata.file_path"))
  .withColumn("ingest_ts", F.current_timestamp())
)
read_data_from_salesdetails.display()
```

##### Write Sales details from DF to bronzre_sales_details as Delta table 

```python
(read_data_from_salesdetails
        .write.mode("overwrite")
        .format('delta')
        .saveAsTable('dlh.bronze_db.bronze_sales_details'))
```
