## Building Bronze Customer AZ12

##### Spark Read Raw Customer AZ12 from Landing CSV file and Create a new DataFrame

```python
from pyspark.sql import functions as F

base_path ='/Volumes/dlh/landing_db/raw_sales/source_erp/CUST_AZ12.csv'

read_data_from_loca101= (
    spark.read
    .option("header","true")
    .option("inferSchema","true")
    .csv(base_path)
    .withColumn("file_path", F.col("_metadata.file_path"))
    .withColumn("ingest_ts", F.current_timestamp())
)
```

##### Write Customer Az12 from DF to bronzre_cust_az12 as Delta table 

```python
(read_data_from_loca101
        .write.mode('overwrite')
        .format('delta')
        .saveAsTable('dlh.bronze_db.bronze_cust_az12'))
```
