## Building Bronze PX Category GV12

##### Spark Read Raw Px category GV12 from Landing CSV file and Create a new DataFrame

```python
from pyspark.sql import functions as F

base_path = '/Volumes/dlh/landing_db/raw_sales/source_erp/PX_CAT_G1V2.csv'

read_from_px_cat_g1v2 = (
    spark.read
    .option("header","true")
    .option("inferSchema","true")
    .csv(base_path)
    .withColumn("file_path", F.col("_metadata.file_path"))
    .withColumn("ingest_ts", F.current_timestamp())

)
```

##### Write PX Category G1V2 from DF to bronzre_px_cat_g1v2 as Delta table 

```python
(read_from_px_cat_g1v2
                    .write
                    .mode("overwrite")
                    .format("delta")
                    .saveAsTable("dlh.bronze_db.bronze_px_cat_g1v2"))
```
