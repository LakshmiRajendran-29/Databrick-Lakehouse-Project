## Silver to Gold Transformation

#####Dimension **Products**

#####Initialization

```python
from pyspark.sql.functions import col, when, row_number
from pyspark.sql import functions as F
from pyspark.sql.window import Window
```

```python
df = spark.table('dlh.silver_db.silver_prod_info')
df.display()
```

```python
df.filter(col("product_id").isNull()).display()
```

```python
df_px = spark.table('dlh.silver_db.silver_px_cat_g1v2')
df_px.display()
```

```python
df_px.filter(col("category_id").isNull()).display()
```

```python
df_join = ( 
           df.alias("p")
                .join(df_px.alias("px"),
                      col("p.category_id") == col("px.category_id"), "left")
                .select(col("p.product_id"),
                        col("p.category_id"),
                        col("p.product_number"),
                        col("p.product_name"),
                        col("p.product_cost"),
                        col("p.product_line"),
                        col("p.start_date"),
                        col("p.end_date"),
                        col("px.category"),
                        col("px.subcategory"),
                        col("px.maintenance_flag")
                        )
)
```

```python
df_join.count()
```

```python
df_filter = df_join.filter(col("p.end_date").isNull())
df_end_dt = df_filter.drop(col("p.end_date"))
df_end_dt.display()
```

```python
df_dup = ( 
          df_end_dt 
            .groupBy("product_id") 
            .agg(F.count("*").alias("cnt")) 
            .filter(col("cnt") > 1)
             )
df_dup.display()
```

```python
from pyspark.sql.functions import col, row_number
from pyspark.sql.window import Window

windowspec = Window.orderBy( col("p.start_date") , col("p.product_number"))

df_pk = df_end_dt.withColumn("product_key", row_number().over(windowspec))
df_pk.display()
```

```python
df_final = df_pk.select(
                 col("product_key"),
                 col("product_id"),
                 col("product_number"),
                 col("product_name"),
                 col("category_id"),
                 col("category"),
                 col("subcategory"),
                 col("maintenance_flag"),
                 col("product_cost"),
                 col("product_line"),
                 col("start_date")
                                              
)
```

```python
df_final.display()
```

```python
spark.sql("DROP TABLE IF EXISTS dlh.gold_db.gold_dim_products")

(
    df_final.write
    .mode("overwrite")
    .format("delta")
    .saveAsTable("dlh.gold_db.gold_dim_products")
)
```
